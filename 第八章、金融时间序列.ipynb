金融中通常使用对数回报率

DataFrame.shift(periods=1, freq=None, axis=0)

    periods可以理解为移动幅度的次数，shift默认一次移动1个单位，也默认移动1次(periods默认为1)，则移动的长度为1 * periods。
    periods可以是正数，也可以是负数。负数表示前滞，正数表示后滞。
    freq是一个可选参数，默认为None，可以设为一个timedelta对象。适用于索引为时间序列数据时。
    freq为None时，移动的是其他数据的值，即移动periods*1个单位长度。
    freq不为None时，移动的是时间序列索引的值，移动的长度为periods * freq个单位长度。
    axis默认为0，表示对列操作。如果为行则表示对行操作。

```python
rets = np.log(data / data.shift(1))
rets.head().round(3)
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(12, 6))       # 绘制一段时间的累计对数回报率图表；首先调用cumsum(), 然后对结果应用np.exp()
```

### 滚动统计

https://zhuanlan.zhihu.com/p/620376722

```python
sym = 'AAPL.O'
data = pd.DataFrame(data[sym]).dropna()
data.tail()
```

```python
window = 20
data['min'] = data[sym].rolling(window=window).min()
data['mean'] = data[sym].rolling(window=window).mean()
data['std'] = data[sym].rolling(window=window).std()
data['media'] = data[sym].rolling(window=window).median()
data['max'] = data[sym].rolling(window=window).max()
data['ewma'] = data[sym].ewm(halflife=0.5, min_periods=window).mean()       # 计算指数加权移动平均值，半衰期为0.5
    # halflife参数决定了指数加权移动平均的速度，其值越大，速度越慢，反之亦然。min_periods参数指定了计算平均值所需的最小观察值数量
```

```python
data.dropna()

ax = data[['min', 'mean', 'max']].iloc[-300:].plot(figsize=(10,6), style=['g--', 'r-.', 'g-.'], linewidth=0.8)

data[sym].iloc[-300:].plot(ax=ax, linewidth=2.0)        # 在图表中添加原始时间序列数据，用ax=ax在同一图表上添加了另一列数据。      
```

```python
data['SMA1'] = data[sym].rolling(window=42).mean()              # 短期SMA
data['SMA2'] = data[sym].rolling(window=252).mean()             # 长期SMA

data[[sym, 'SMA1', 'SMA2']].plot(figsize=(10,6))
```

```python
data.dropna(inplace=True)

data['positions'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)

ax = data[[sym, 'SMA1', 'SMA2', 'positions']].plot(figsize=(10,6), secondary_y='positions')     # secondary_y'参数用于在图形上创建第二个y轴
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))                # 通过set_bbox_to_anchor方法，你可以指定图例的锚点位置，即图例框在图形中的位置
```

对数回报率

pandas scatter_matrix() 可绘制两个序列对数回报率的对比，用户可在对角线上添加直方图或核密度估计量（KDE）

pd.plotting.scatter_matrix是一个绘制散点矩阵图的函数。它接受一个DataFrame作为输入，并为其中的每对列创建一个散点图矩阵。它主要用于显示DataFrame中各列之间的相关性。

```python
pd.plotting.scatter_matrix(rets, alpha=0.2, diagonal='hist', hist_kwds={'bins':35}, figsize=(10,6))
    # diagonal：设置对角线上的图。可选值包括'hist'（直方图）和None。在这个例子中，diagonal设置为'hist'，意味着在每个对角线上绘制一个直方图。
    # hist_kwds：用于设置直方图的参数。在这个例子中，设置'bins'参数为35，表示将数据分成35个等间距的直方图条形
```

OLS回归

```python
reg = np.polyfit(rets['.SPX'], rets['.VIX'], deg=1)         
    # 实施线性OLS回归
    # polyfit函数对两个时间序列数据进行线性拟合。这个函数返回一个表示拟合多项式的数组，你可以用它来评估拟合线的值。
    # 这里的deg=1表示你正在进行一次线性拟合，也就是拟合一个直线。

ax = rets.plot(kind='scatter', x='.SPX', y='.VIX', figsize=(10,6))

ax.plot(rets['.SPX'], np.polyval(reg, rets['.SPX']), 'r', lw=2)     
    # 添加线性回归线，np.polyval(reg, x)是NumPy库中的一个函数，它根据给定的多项式系数（reg）和输入值（x）来计算多项式的值
```
