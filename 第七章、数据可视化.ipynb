```python
def func(x):
    return 0.5 * np.exp(x) + 1

a, b = 0.5, 1.5                                         # 积分上下限
x = np.linspace(0, 2)                                   # 图表x值
y = func(x)                                             # 积分上下限的x值
Ix = np.linspace(a, b) 
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]         # 列表对象，包含表示所绘制多边形坐标的元组对象
```

```python
from matplotlib.patches import Polygon

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'g', linewidth=2)
plt.ylim(bottom=0)                      # 为纵轴定义最小y值
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')         # 使用Polygon生成阴影部分，将多边形（积分区域）绘制成灰色
ax.add_patch(poly)
plt.text(0.5 * (a+b), 1, r'$\int_a^b f(x)\mathrm{d}x$', horizontalalignment='center', fontsize=20)      # 在图中放置积分公式
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

ax.set_xticks((a,b))                    # 设置了x轴的刻度
ax.set_xticklabels(('$a$', '$b$'))      # 设置了x轴的标签
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
```

```python
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
import numpy as np

# 定义一个求积分的函数

def func(x):
    return 0.3* (x**2) + (0.1*x) + 1

# 定义积分区间

a, b = 1,2
x = np.linspace(0,3)
y = func(x)

# 绘制曲线

fig, ax = plt.subplots(figsize=(10,6))
plt.plot(x, y, 'g')
plt.ylim(ymin=0)

# 使用Polygon生成阴影部分

Ix = np.linspace(a, b)
Iy = func(Ix)

# verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
verts = [(a,0)] + list(zip(Ix, Iy))                 # 多边形的顶点列表verts
ploy = Polygon(verts, facecolor = 'orange', edgecolor='0.5', alpha=0.6)

ax.add_patch(ploy)              # matplotlib库的axies模块中的Ax.add_patch()函数用于将补丁添加到轴的补丁；返回补丁。将这个多边形添加到了图形上。

# 添加数学公式和坐标轴标签

# r" "中间为LaTex语法表示的公式

plt.text(0.5 * (a+b), 1, r"$\int_a^b f(x)\mathrm{d}x$",
        horizontalalignment='center', fontsize=20)

# 前两个参数是放置文本的坐标

plt.figtext(0.9, 0.075,'$x$')
plt.figtext(0.075,0.9,'$f(x)$' )            # $x$ 能使得其中的字符变为斜体

ax.set_xticks((a,b))
ax.set_xticklabels(('a','b'))
# 第一行代码ax.set_xticks((a, b))设置x轴的刻度位置，a和b是你想要放置刻度的位置。
# 第二行代码ax.set_xticklabels(('$a$','b'))设置x轴的刻度标签，'$a$'和'b'是对应的标签。你需要确保变量a和b在你调用这些函数时已经被定义。

ax.set_yticks((func(a), func(b)))
ax.set_yticklabels(('f(a)','f(b)'))

plt.grid(True)
plt.show()
```

### 静态3D绘图

numpy.meshgrid()理解

直观理解：meshgrid() 用于生成网格采样点矩阵



https://blog.csdn.net/lllxxq141592654/article/details/81532855

https://zhuanlan.zhihu.com/p/29663486

```python
strike = np.linspace(50, 150, 24)
ttm = np.linspace(0.5, 2.5, 24)

strike, ttm = np.meshgrid(strike, ttm)      # 创建的两个二维ndarray对象（网格）
                                            # np.meshgrid函数接受两个一维数组，表示网格的X轴和Y轴，并返回两个二维数组表示对应的X轴和Y轴网格点
strike[:2].round(1)

iv = (strike - 100) ** 2 / (100 * strike) /ttm          # 虚拟的隐含波动率值

iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
# ax = fig.gca(projection='3d')         # FigureBase.gca() got an unexpected keyword argument 'projection', Axes3D函数没有用到，报错       
ax = fig.add_axes(Axes3D(fig))          # 创建3D图表

surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2, cmap=plt.cm.coolwarm, linewidth=0.5, antialiased=True)        # plot_surface函数通常用于绘制三维数据场的表面图
        # rstride和cstride是matplotlib库中3D图像函数plot_surface()的参数。其值最小为1，最大可以无穷大。当rstride和cstride的值为1时，图像不会变化；
        # 增加rstride和cstride的值，会减少三维图像的平滑程度；当其等于X或Y的默认栅格数时，就变成一个单色面
        # antialiased=True, 图形将会进行抗锯齿处理，边缘会更加平滑

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=10)        # 用于添加颜色条的函数
    # shrink参数是用来设置颜色条相对于图像的大小。0.5表示颜色条的宽度是图像宽度的50%。
    # aspect参数是用来设置颜色条的高度与宽度比。在你提供的代码中，这个值设置为10。这意味着颜色条的高度是其宽度的10倍。

# ax.view_init(45, 45)                  # 选择视角
```

```python
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)
ax.scatter(strike, ttm, iv, zdir='z', s=25, c='b', marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
```

```python
import matplotlib.pyplot as plt
from numpy import *
from mpl_toolkits.mplot3d import Axes3D

# 设置坐标轴
fig = plt.figure(figsize=(15,6))      # 定义一个窗口
ax = fig.add_axes(Axes3D(fig))        # 在窗口添加一个3d坐标轴
 
# 设置X,Y,Z的值
X = arange(-4,4,0.25)                 # 范围是[-4,4],步长为0.25,生成一个数组
Y = arange(-4,4,0.25)
X,Y = meshgrid(X,Y)                   # 生成绘制3D图形所需要的网格数据
R = sqrt(X**2+Y**2)
Z = sin(R)
 
# 将图画在3d坐标上
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'), edgecolor='black')       # rstride:行跨，cstride:列跨
# 将3d图像映射到xoy轴
ax.contourf(X, Y, Z, zdir='Z', offset=-2, cmap='rainbow')        # zdir='Z':将z轴压到xoy平面上，offset=-2:将图形映射到z=-2的平面上
 
ax.set_zlim(-2,2)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x,y = np.mgrid[-5:5:200j,-5:5:200j]
sigma = 2
z = 1/(2 * np.pi * (sigma**2)) * np.exp(-(x**2+y**2)/(2 * sigma**2))


fig = plt.figure()
ax = fig.add_axes(Axes3D(fig)) 
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='rainbow',alpha = 0.9)

plt.show()
```

添加典型的金融图表元素——布林线

    布林线(BOLL)，Bollinger Bands，利用统计原理，求出股价的标准差及其信赖区间，从而确定股价的波动范围及未来走势，利用波带显示股价的安全高低价位，因而也被称为布林带

```python
qf.add_bollinger_bands(periods=15, boll_std=2)
    # periods 设定布林线周期
    # 用于布林线宽度的标准差

plyo.iplot(qf.iplot(asFigure=True))
```
